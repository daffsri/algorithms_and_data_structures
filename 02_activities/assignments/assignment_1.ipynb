In other words, we need to implement a function that checks if any value appears more than once. If multiple values appear multiple times, we are interested only in the one that appears first as determined by distance from the root.

```python
from collections import deque

class TreeNode(object):
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

def is_duplicate(root: TreeNode) -> int:
    if not root:
        return -1

    seen_values = set() 
    queue = deque([root]) 

    while queue:
        node = queue.popleft() 

        if node.val in seen_values:
            return node.val
        
        seen_values.add(node.val)
        queue.append(node.left) if node.left else None
        queue.append(node.right) if node.right else None

    return -1
```


-   Explain why your solution works


| This solution works by using a breadth-first search (BFS) to traverse the tree starting from the root, level by level. It keeps track of the values of each node it encounters using a set, as well as a queue of child nodes of the next level. This allows for efficient duplicate check to ensure the first duplicate found and returned will be the one closest to the root. When a duplicate is found, the function returns that value immediately. If no duplicates are found, it returns `-1`.


-   Explain the problem’s time and space complexity


| Where `n` is the number of nodes:

`Time complexity: O(n)` --> each node is traversed once at most, and operation of constant time is perfomed at each node, therefore total time complexity is n

`Space complexity: O(n)` --> recursion stack, if unbalanced worst case scenario is n. Set size, worse case scenario would be of size n. Therefore space complexity in worst case scenario of 2n = n.
