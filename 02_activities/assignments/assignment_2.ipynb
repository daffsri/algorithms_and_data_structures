explantion: given the list [0, 4] then the range is 0 to 4 (ie: 0, 1, 2, 3, 4). However the input list only has 0 and 4 therefore the remaining integers are "missing" and will be returned as the missing list [1, 2, 3]
    



-   Copy the solution your partner wrote. 


```python
# Partner's solution
from typing import List, Union

def missing_num(nums: List[int]) -> Union[int, List[int]]:
    if not nums:
        return -1

    # Define n as the maximum number in the list, which represents the end of the range [0, n]
    n = max(nums)
    
    # Convert list to set for O(1) lookups
    num_set = set(nums)
    
    # Build a list of numbers in the range [0, n] that are missing from nums
    missing = [i for i in range(n + 1) if i not in num_set]
    
    # If no numbers are missing, return -1
    return missing if missing else -1

# Examples:
print(missing_num([0, 2]))                   
print(missing_num([5, 0, 1]))                  
print(missing_num([6, 8, 2, 3, 5, 7, 0, 1, 10]))  
print(missing_num([0, 1, 1, 1, 1]))
```


-   Explain why their solution works in your own words.


| Given a list on n integers based on the assumption the values of this list will be restricted to [0, n] the missing function does the following:
* if input array is empty list - return `-1`
* if non empty array set `n = max of array` (mistake)
* loop through `0 to n` (range of n+1 will stop at n) to find any missing numbers using list comprehension
* return list of missing if missing is not an empty array, else return `-1`


-   Explain the problem’s time and space complexity in your own words.


Given `m` length of missing array and `n` length of input array

| space complexity: `O (m + n)`

* O (m): Array of missing can be at most m (due to the incorrect way n is calculated by max) 
* O (n): Set of space n from the n input array items. 

| time complexity: `O (m + n)`

* O (n): to iterate through input array and calculate max
* O (n): to put the input array into a set
* O (1): constant time to lookup in set 
* O (m): to build the missing number array of m items

The correct optimized solution would have time and space complexity of `O(n)`
